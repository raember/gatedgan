```python
from pathlib import Path
from pdf2image import convert_from_path
```

```python
imslp_path = Path('../imslipper')
rev_sorted_list_path = imslp_path / 'reverse_sorted_size.list'
out_folder = Path('./imslp')
NSAMPLES = 50
NSAMPLESPERFILE = 30
```

```python
out_folder.mkdir(exist_ok=True)
```

```python
processed = []
with open(rev_sorted_list_path, 'r') as fp:
    for _ in range(NSAMPLES):
        file_path = imslp_path / fp.readline().lstrip('0123456789').strip()
        if file_path.name in processed or file_path.name.startswith('IMSLP563117'):
            continue
        processed.append(file_path.name)
        print(f"Processing {file_path.name}")
        im = convert_from_path(
            file_path,
            output_folder=out_folder,
            first_page=1,
            last_page=NSAMPLESPERFILE,
            fmt="jpeg",
            output_file=f"{file_path.name}"
        )
```
